# Importing the Required modules

```python
import easygui
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imageio
```

# Setting the path of the Image

```python
# Here we provide the path of the image that is to be used
img = "1 (1009).jpg" 
```

# Loading the Image in a variable

```python
# Here we use the "imread()" function of cv2 to read the image whose path stored in the variable "img", and store it in "img1"
# The image is stored in the form of matrix 
img1 = cv2.imread(img) 
print(img1)
```

```python
# Here we use the Matplotlib module to display the image
# The "imshow()" function converts the matrix stored in "img1" to image form
plt.imshow(img1, cmap = "gray")
```

# Converting the IMage from BGR TO RGB

```python
# opencv stores the image in "BGR" form by default
# as a result the human eyes gets to see an inverted color image
#using "cvtColor()" function we transform the image from "BGR" to "RGB" form
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
print(img1)
```

```python
plt.imshow(img1)
```

# Convert Image from RGB to Grayscale

```python
# The first step to cartoonify an Image is to convert it into grayscale
img3 = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
print(img3)
```

```python
plt.imshow(img3, cmap = "gray")
```

# Smoothening of the Grayscale Image

```python
# The next step is to smoothen the image
# this is done by using the "medianBlur()" function to blur the image 
smoothimg = cv2.medianBlur(img3,5)
print(smoothimg)
```

```python
plt.imshow(smoothimg, cmap = "gray")
```

# Retrieving the edges(boundary) of the object in the image

```python
# The next step is to get the boundaries/edges of the image 
# This is done by using the "adaptiveThreshold()" function of cv2
edgeimg = cv2.adaptiveThreshold(smoothimg, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 9, 9)
print(edgeimg)
```

```python
plt.imshow(edgeimg, cmap = "gray")
```

# Making a lightened Color Image of the original image 

```python
# In the next step we create a lightened image of the original image 
# This is done by using the "bilateralFilter()" function of cv2
colorimg = cv2.bilateralFilter(img1, 9, 300, 300)
print(colorimg)
```

```python
plt.imshow(colorimg)
```

# Merging the edge image with the Lightened color Image

```python
# The last step is to merge the smoothened and the edged image to get the cartoonified image
# This is done by using the "bitwise_and()" function of cv2 
cartoonimg = cv2.bitwise_and(colorimg, colorimg, mask = edgeimg)
print(cartoonimg)
```

# THE CARTOONIFIED IMAGE

```python
plt.imshow(cartoonimg)
```
